```python
def mapper(matrix_A, matrix_B, m, n):
    mapped_results = []

    # Emit key-value pairs for matrix A
    for i in range(m):
        for j in range(n):
            mapped_results.append(((i, j), ("A", matrix_A[i][j])))

    # Emit key-value pairs for matrix B
    for i in range(m):
        for j in range(n):
            mapped_results.append(((i, j), ("B", matrix_B[i][j])))

    return mapped_results

def shuffle_and_sort(mapped_results):
    shuffled_data = {}

    # Group values by key (i, j)
    for key, value in mapped_results:
        if key in shuffled_data:
            shuffled_data[key].append(value)
        else:
            shuffled_data[key] = [value]

    return shuffled_data

def reducer(shuffled_data):
    result_matrix = {}

    # Compute addition for each key (i, j)
    for key, values in shuffled_data.items():
        i, j = key
        A = 0
        B = 0

        # Get values from A and B
        for value in values:
            matrix, val = value
            if matrix == "A":
                A = val
            elif matrix == "B":
                B = val

        # Perform matrix addition
        result_matrix[(i, j)] = A + B

    return result_matrix

if __name__ == "__main__":
    matrix_A = [
        [1, 2],
        [3, 4]
    ]

    matrix_B = [
        [5, 6],
        [7, 8]
    ]

    m = len(matrix_A)      # Rows
    n = len(matrix_A[0])   # Columns

    mapped_results = mapper(matrix_A, matrix_B, m, n)

    shuffled_data = shuffle_and_sort(mapped_results)

    result_matrix = reducer(shuffled_data)

    print("Resultant Matrix C:")
    for (i, j), value in result_matrix.items():
        print(f"C[{i}][{j}] = {value}")
```
