```python
"""4. Design and implement a CNN for Image Classification a) Select a suitable image classification 
dataset (medical imaging, agricultural, etc.). b) Optimized with different hyper-parameters including 
learning rate, filter size, no. of layers, optimizers, dropouts, etc."""

import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
```

```python
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.cifar10.load_data()
```

```python
# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0
```

```python
LEARNING_RATE = 0.001
FILTER_SIZE = (3, 3)
DROPOUT_RATE = 0.3
OPTIMIZER = optimizers.Adam(learning_rate=LEARNING_RATE)
```

```python
def build_model():
    model = models.Sequential()
    # Layer 1: Convolution + Pooling
    # Increasing 'no. of layers' or 'filter count' (32, 64) changes complexity
    model.add(layers.Conv2D(32, FILTER_SIZE, activation='relu', input_shape=(32, 32, 3)))
    model.add(layers.BatchNormalization())
    model.add(layers.MaxPooling2D((2, 2)))
    # Layer 2: Deeper Convolution
    model.add(layers.Conv2D(64, FILTER_SIZE, activation='relu'))
    model.add(layers.Dropout(DROPOUT_RATE)) # Dropout to prevent overfitting
    model.add(layers.MaxPooling2D((2, 2)))
    # Layer 3: Deepest Convolution
    model.add(layers.Conv2D(64, FILTER_SIZE, activation='relu'))
    # Flattening for the Dense layers
    model.add(layers.Flatten())
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(10)) # 10 classes in CIFAR-10
    return model
model = build_model()
```

```python
# Compile with chosen optimizer
model.compile(optimizer=OPTIMIZER,
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(train_images, train_labels, epochs=10,
                    validation_data=(test_images, test_labels))
```
