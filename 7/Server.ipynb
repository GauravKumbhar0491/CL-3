```python
from xmlrpc.server import SimpleXMLRPCServer
server = SimpleXMLRPCServer(("localhost", 1233))

@server.register_instance
class StringService:
    #@staticmethod                   # OPTIONAL you can safely delete this line. Just there to respect the syntax
    def concat(a: str, b: str) -> str:
        return a+b

print("RPC Server running on port 1234...")
server.serve_forever()
```
