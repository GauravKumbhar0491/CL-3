```python
"""Implement Union, Intersection, Complement and Difference operations on fuzzy sets. Also
create fuzzy relations by Cartesian product of any two fuzzy sets and perform max-min
composition on any two fuzzy relations."""

A = {'x1':0.2, 'x2':0.7, 'x3':1.0}
B = {'x1':0.6, 'x2':0.4, 'x3':0.8 }
C = {'y1':0.5, 'y2':0.9}
```

```python
def union(A, B):
    return{x: max(A[x], B[x]) for x in A}

print("Union:", union(A, B))
```

```python
def intersection(A, B):
    return{x: min(A[x], B[x]) for x in A}

print("Intersection:", intersection(A, B))
```

```python
def Difference(A,B):
    return{x: min(A[x], 1 - B[x]) for x in A}

print("Difference:", Difference(A, B))
```

```python
def Complement(A):
    return{x: 1 - A[x] for x in A}

print("Complement:", Complement(A))
```

```python
def cartesion_product(A, B):
    relation = {}
    for x in A:
        for y in B:
            relation[(x, y)] = min(A[x], B[y])
    return relation

print("Fuzzy Relation :", cartesion_product(A, B))
```

```python
def max_min_composition(R1, R2, X, Y, Z):
    result = {}
    for x in X:
        for z in Z:
            max_val = 0
            for y in Y:
                if (x, y) in R1 and (y, z) in R2:
                    value = min(R1[(x, y)], R2[(y, z)])
                    max_val = max(max_val, value)
            result[(x, z)] = max_val
    return result

R1 = cartesion_product(A, B)
R2 = cartesion_product(B, C)
X = list(A.keys())
Y = list(B.keys())
Z = list(C.keys())

print("Max-Min Composition :", max_min_composition(R1, R2, X, Y, Z))
```
