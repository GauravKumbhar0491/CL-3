```python
#Import Data from different Sources such as (Excel, Sql Server, Oracle etc.) and load in targeted system.


import pandas as pd
import sqlite3
from sqlalchemy import create_engine

print("Libraries loaded successfully")

csv_data = {
"customer_id": [1,2,3],
"name": ["John","Alice","Rahul"],
"city": ["Pune","Mumbai","Delhi"],
"purchase_amount": [200,350,500]
}

df_csv = pd.DataFrame(csv_data)

df_csv.to_csv("customer_data.csv", index=False)

print("\nCSV Source Data:")
print(df_csv)

excel_data = {
"customer_id": [4,5,6],
"name": ["Sara","David","Neha"],
"city": ["Bangalore","Chennai","Hyderabad"],
"purchase_amount": [450,300,600]
}

df_excel = pd.DataFrame(excel_data)

df_excel.to_excel("customer_data.xlsx", index=False)

print("\nExcel Source Data:")
print(df_excel)
conn = sqlite3.connect("source_database.db")
sql_data = {
"customer_id": [7,8,9],
"name": ["Rohit","Priya","Amit"],
"city": ["Nagpur","Jaipur","Surat"],
"purchase_amount": [250,420,390]
}

df_sql = pd.DataFrame(sql_data)

df_sql.to_sql("customer_table", conn, if_exists="replace", index=False)

print("\nSQL Source Data:")
print(df_sql)

csv_source = pd.read_csv("customer_data.csv")
excel_source = pd.read_excel("customer_data.xlsx")
sql_source = pd.read_sql("SELECT * FROM customer_table", conn)

print("\nExtracted CSV Data:")
print(csv_source.head())

print("\nExtracted Excel Data:")
print(excel_source.head())

print("\nExtracted SQL Data:")
print(sql_source.head())

combined_data = pd.concat([csv_source, excel_source, sql_source], ignore_index=True)

combined_data = combined_data.drop_duplicates()

print("\nTransformed Combined Data:")
print(combined_data)

engine = create_engine("sqlite:///target_database.db")

combined_data.to_sql(
name="final_customer_data",
con=engine,
if_exists="replace",
index=False
)

print("\nData successfully loaded into Target Database!")

result = pd.read_sql("SELECT * FROM final_customer_data", engine)

print("\nData in Target Database:")
print(result)

print("\nETL Process Completed Successfully")
```
