```python
#Perform the Extraction Transformation and Loading (ETL) process to construct the database in the Sql server / Power BI.




import pandas as pd
import sqlite3
```

```python
data = pd.read_csv("uber.csv")

print("Extracted data")
print(data.head())
```

```python
data = data.drop_duplicates()

data = data.fillna(method='ffill')


print("Transformed Data")
print(data.head())
```

```python
conn = sqlite3.connect("BI_database.db")

data.to_sql("uber_table",conn,if_exists="replace",index=False)

print("Data loaded into Target Database")
```

```python
result = pd.read_sql("SELECT * FROM uber_table",conn)

print(result.head())
```
